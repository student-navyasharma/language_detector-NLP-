```python
from langdetect import detect
```

```python
text="Bonjour"
language=detect(text)
print(language)
```

```python
language_map={
     "en": "English",
    "fr": "French",
    "hi": "Hindi",
    "es": "Spanish",
    "de": "German",
    "it": "Italian",
    "pt": "Portuguese",
    "nl": "Dutch"
}

text="Bonjour"
lang_code=detect(text)
print(language_map.get(lang_code,"Unknown"))
```

```python
text=input("Enter a text")
lang_code=detect(text)
print(language_map.get(lang_code,"Unknown"))
```

```python
from langdetect import detect, LangDetectException

try:
    text = input("Enter text: ")
    lang_code = detect(text)
    print("Language:", language_map.get(lang_code, "Unknown"))

except LangDetectException:
    print("Text too short or unclear!")
```
